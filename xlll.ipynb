```python
import xml.etree.ElementTree as ET

# XML data as a string (replace this with your actual XML content)
xml_data = """
==[[漢字]]==
&lt;span lang=&quot;ja&quot; xml:lang=&quot;ja&quot; style=&quot;font-size:350%&quot;&gt;{{PAGENAME}}&lt;/span&gt;
*{{部首|木|10}}
*{{総画|14}}
*[[異体字]] : &lt;span style=&quot;font-size:250%&quot;&gt;[[槪]]&lt;/span&gt;（[[旧字体]]）, &lt;span style=&quot;font-size:250%&quot;&gt;{{JIS2004font|&amp;#x69EA;}}&lt;/span&gt;（異体字）, &lt;span style=&quot;font-size:250%&quot;&gt;[[槩]]&lt;/span&gt;（異体字）
*筆順 : [[image:{{PAGENAME}}-bw.png|350px]]
===字源===
* [[形声文字|形声]]。「[[木]]」+音符「[[既]] {{phoneme|*KƏT}}」。「[[とかき]]」を意味する[[漢語]]｛[[概]] {{phoneme|*kəəts}}｝を表す字。

===意義===
#[[とかき]]、枡のふちを超えた部分を[[すりきる|すりきっ]]て、ならすための小型の棒板。
#[[おおむね]]。
[[Category:漢字]]
[[Category:常用漢字]]

=={{ja}}==
[[Category:{{ja}}]]
==={{pron|ja}}===
* 音読み
** [[呉音]] : [[カイ]]
** [[漢音]] : [[カイ]]
** [[慣用音]] : [[ガイ]]
* 訓読み
** 常用漢字表内
**: 常用漢字表内の訓読みはありません。
** 常用漢字表外
**: [[おおむね|おおむ-ね]]、[[とかき]]
==={{noun}}===
[[Category:{{ja}} {{noun}}|とかき]]
#（[[とかき]]）枡からはみ出た分をならす棒（「おおむね」の原義）

==={{drv}}===
*[[概して]]、[[慨する]]

==={{prov}}===
{{top}}
*[[概評]]
*[[概数]]
*[[概説]]
*[[概観]]
*[[概算]]
*[[概念]]
*[[概容]]
*[[概覧]]
*[[概略]]
*[[概要]]
*[[気概]]
*[[大概]]
{{bottom}}

=={{zh}}==
[[Category:{{zh}}|gai4]]
[[Category:常用字|gai4]]
{{trans_link|zh|概}}
* '''ローマ字表記'''
** '''[[普通話]]'''
*** '''[[ピンイン]]''': gài (gai4)
*** '''[[ウェード式]]''': kai&lt;sup&gt;4&lt;/sup&gt;
*** '''[[注音符号]]''': ㄍㄞˋ
** '''[[広東語]]'''
*** '''[[香港言語学学会式]]''': koi2, koi3
*** '''[[イェール式]]''': koi2, koi3
==={{prov}}===

=={{ko}}== 
[[Category:{{ko}}|ㄱㅐ]] 

{{trans_link|ko|{{PAGENAME}}}}

* '''[[ハングル]]''': {{lang|ko|[[개]]}}
* '''ローマ字表記'''
** [[文化観光部2000年式]]: gae
** [[マッキューン＝ライシャワー式]]: kae
** [[イェール式]]: kay

=={{vi}}==
[[Category:{{vi}}|khai3 khai]]
{{trans_link|vi|{{PAGENAME}}}}
* '''ローマ字表記'''
** [[Quốc ngữ]]: [[khái]]
"""

# Function to extract and print entire sections
def extract_sections(xml_content, sections):
    # Split the content by lines
    lines = xml_content.split('\n')
    
    # Variable to keep track of whether we are within a desired section
    in_section = False
    section_lines = []  # List to hold the lines of the current section
    current_section_level = 0  # Track the level of the current section

    # Helper function to determine the level of the heading
    def heading_level(line):
        return line.count('=')

    # Iterate over each line
    for line in lines:
        # Check if the line is a heading
        if line.startswith('=='):
            # Determine the level of the heading
            level = heading_level(line.strip())
            
            # Check if we are starting a new section
            if any(line.strip() == f"{'=' * level}{section}{'=' * level}" for section in sections):
                if in_section:
                    # We've reached a new section, so print and clear the current section
                    print('\n'.join(section_lines))
                    section_lines = []
                in_section = True  # Start capturing the lines for this section
                current_section_level = level  # Update the current section level
                section_lines.append(line)  # Include the heading in the section
            elif in_section and level <= current_section_level:
                # We've reached the end of the section or a new section at the same level
                in_section = False
                print('\n'.join(section_lines))
                section_lines = []
                current_section_level = 0  # Reset the section level
        elif in_section:
            # We're still in the section, so add the line to the list
            section_lines.append(line)

    # Print the last section if we haven't already
    if in_section:
        print('\n'.join(section_lines))

# Define the sections you want to extract
sections_to_extract = ['{{prov}}', '{{zh}}', '{{ja}}', '{{ko}}', '{{vi}}', '{{drv}}']

# Call the function with the XML content and the sections to extract
extract_sections(xml_data, sections_to_extract)

```
